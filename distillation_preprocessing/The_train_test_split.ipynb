```python
import json
import random
import os

FILES_TO_SPLIT = [
    {"path": "phase1_enron_clean.jsonl", "split_ratio": 0.90},
    {"path": "phase1_sroie_input.jsonl", "split_ratio": 0.85},
    {"path": "phase1_ami_clean.jsonl", "split_ratio": 0.85},
    # FUNSD is already split by folders
]

def split_dataset(file_info):
    input_path = file_info["path"]
    train_ratio = file_info["split_ratio"]
    
    if not os.path.exists(input_path):
        print(f"Skipping {input_path} (not found)")
        return

    print(f"Splitting {input_path}...")
    
    with open(input_path, 'r', encoding='utf-8') as f:
        lines = f.readlines()
    
    # Shuffle to ensure random distribution
    random.seed(42) # Fixed seed for reproducibility
    random.shuffle(lines)
    
    split_index = int(len(lines) * train_ratio)
    train_data = lines[:split_index]
    test_data = lines[split_index:]
    
    # Define new filenames
    base_name = input_path.replace(".jsonl", "")
    train_file = f"{base_name}_train.jsonl"
    test_file = f"{base_name}_test.jsonl"
    
    # Save Train
    with open(train_file, 'w', encoding='utf-8') as f:
        f.writelines(train_data)
        
    # Save Test
    with open(test_file, 'w', encoding='utf-8') as f:
        f.writelines(test_data)
        
    print(f"  -> Train: {len(train_data)} samples saved to {train_file}")
    print(f"  -> Test:  {len(test_data)} samples saved to {test_file}")

for file_info in FILES_TO_SPLIT:
    split_dataset(file_info)

print("\n" + "="*30)
print("Splitting Complete.")
```
